## LA fonction apply permet d'appliquer une function sur chaque élément d'une série => exercice : appliquer apply pour transformer les éléments de la colonne 'long' en float

```python
new_long = accidents.long.apply(lambda x: float(x) if "E" not in str(x) else None)
new_long.value_counts().head()
```

# Remplacer la colonne long par la nouvelle colonne "new_long" 

```python
accidents['long'] = new_long
```
