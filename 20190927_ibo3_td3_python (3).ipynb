# 3.9) Préciser l'encoding du fichier : ici latin-1, ce qui n'est pas la norme attendue par python, qui veut de l'UTF-8

# 3.10) Exercice : créer deux colonnes, une avec l'heure, l'autre avec les minutes

```python
hours = accidents2.hrmn.apply(lambda x: f"{x:04}"[:2])
minutes = accidents2.hrmn.apply(lambda x: f"{x:04}"[2:])
```

```python
hours = accidents2.hrmn.apply(lambda x: x//100)
minutes = accidents2.hrmn.apply(lambda x: x%100)
```

```python
accidents2["hours"] = hours
accidents2["minutes"] = minutes
```

# 4) groupby

```python
accidents2.groupby(["hours", "minutes"])["Num_Acc"].count().plot(kind="bar")
```

# 5) création d'une colonne de date

# 5.1) on utilise ZIP pour récupérer l'année, le mois, le jour

```python
for an,mois, jour in zip(accidents2.an, accidents2.mois, accidents2.jour):
    print(an,mois, jour)
    break
```

## 5.1 bis) ZIP dans une comprehension list

```python
r = [f'{an:02}{mois:02}{jour:02}' for an,mois, jour \
 in zip(accidents2.an, accidents2.mois, accidents2.jour)]
    
```

```python
r[:3]
```

## 5.2) conversion de la liste de chaîne de caractères en liste d'objets de type datatime grâce à la fonction to_datetime de pandas

```python
nouvelles_dates =  pd.to_datetime(r, format="%y%m%d")
```

## 5.3) rajout des dates dans une nouvelle colonne

```python
accidents2["date_accident"] = nouvelles_dates
```

## 5.4) on applique un "apply" pour récupérer le jour de la semaine via l'attribut 'dayofweek' de l'objet date créé

```python
accidents2["day_of_week"] = accidents2.date_accident.apply(lambda x: x.dayofweek )
```

```python
import seaborn
```

```python
gp = accidents2.groupby(["mois", "day_of_week"])["Num_Acc"].count()
gp.plot(kind="bar")
gp[gp==gp.max()]
seaborn.despine()
```

# 9) tableau croisé dynamiqe

```python
tcd = jointure.pivot_table(          columns=["categorie_passagers"],
                           
                      index="sexe",    values=["Num_Acc"], 
                                       aggfunc="count")
tcd
```

## 9.1) calculer le pourcentage par catégorie

```python
tcd.sum().sum()
```

```python
tcd2 = tcd / tcd.sum().sum()
tcd2
```
